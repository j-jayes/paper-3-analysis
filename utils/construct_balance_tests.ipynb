In this file we construct the balance test dataset

The balance test dataset comes from `More Power to the People`

Here we change the definition of the treatment variable - expand the number of parishes

Then we join our outcome variables, unemployment, income, wealth, gini coefficients for income and wealth. (So we need to find these results first.)

```python
import pandas as pd
import pathlib

root = pathlib.Path.cwd().parent

df = pd.read_stata(root / "data/balance-tests/data Figure 2 and Table A2.dta")

df.columns
```

```python
treated_parishes = pd.read_excel(root / "data/parishes/treated_parishes.xlsx")

treated_parishes["treated"] = treated_parishes["iline"]

treated_parishes = treated_parishes.drop(["iline", "parish_name", "parish_name_lower"], axis=1)

df = df.merge(treated_parishes, how = "left", on = "parish_code")

df["treated"] = df["treated"].fillna(0)
```
